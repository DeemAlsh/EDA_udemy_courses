```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline 

Data = pd.read_csv('udemy_courses.csv')
Data
```

```python
Data.info()
```

```python
Data.columns
```

```python
Data.isna()
```

```python
Data.drop('url',
  axis='columns', inplace=True)
```

```python
Data.rename(columns={
    'course_id': 'Course_Id',
    'course_title': 'Course_Title',
    'is_paid': 'Is_Paid',
    'price': 'Price',
    'num_subscribers': 'Num_Subscribers',
    'num_reviews': 'Num_Reviews',
    'num_lectures': 'Num_Hours',
    'level': 'Level',
    'content_duration': 'Content_Duration',
    'published_timestamp': 'Published_Timestamp',
    'subject ': 'Subject ',
}, inplace=True)
```

```python
Data.columns
```

```python
plt.figure(figsize=(20,10))

plt.subplot(1,2,1)
Data.groupby('subject').Num_Subscribers.count().plot(kind='bar');
plt.title("Number of Courses per Subject ", size=20)
plt.ylabel("Count");
plt.xticks(rotation=0)


plt.subplot(1,2,2)
Data.groupby('subject').Num_Subscribers.sum().plot(kind='bar');
plt.title("Number subscribers of each subject ", size=20)
plt.ylabel("Num_Subscribers");
plt.xticks(rotation=0);
```

```python
Data2 = pd.DataFrame({})
D = Data.subject
S = D.value_counts()
R = Data.groupby('subject').Num_Subscribers.sum()
Data2= pd.DataFrame({'Num Courses': S, 'Subscribers': R})
Data2
```

```python
A= Data['Price'].idxmin()
B= Data['Price'].max()

print("The minimum number of courses price = %d" %(A))
print("The maximum number of courses price = %d" %(B))
```

```python
plt.figure(figsize=(15,10))
TF =['True', 'False']
J = Data['Is_Paid'].value_counts()
color = ["teal", "pink"]
plt.pie(J, labels=TF, colors = color, autopct='%1.1f%%');
plt.title('Paid Free Course Pie Chart', size=20);
```

```python
Data['Is_Paid'].value_counts()
```

```python
plt.figure(figsize=(10,8))
Data.groupby('Level').Num_Subscribers.count().plot(kind='bar')
plt.title('Level vs No.Subscribers', size=20)
plt.ylabel('Num Subscribers');
plt.xticks(rotation=0);
```

```python
Data3 = pd.DataFrame({})
D = Data.Level
S = D.value_counts().sort_values()
R = Data.groupby('Level').Num_Subscribers.sum().sort_values()
Data3= pd.DataFrame({'Num Courses': S, 'Subscribers': R})
Data3
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='subject', y='Price', data=Data, palette='magma');
plt.title('Subject/Courses Prices', size=20);
```

```python
print('Top 5 most expensive Web Development course')

idx = pd.IndexSlice
most = Data.loc[idx[Data[Data['subject']=='Web Development']['Price'].sort_values(ascending=False).index]][['Course_Title', 'Price', 'Num_Subscribers']]
most.head(5)
```
